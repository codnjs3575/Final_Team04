```python
from konlpy.tag import Okt
okt = Okt()
```

```python
kakao_df.iloc[0,2]
```

```python
## 불용어
file = open("/content/drive/MyDrive/data/stop_words.txt", "r", encoding='utf-8')
data = file.read()
file.close()

stopwords = data + 'ㅋㅋ' + 'ㅠㅠㅠ' + 'ㅜㅜ'+'ㅠㅠ'+'ㅎㅎ'+'맛있다'+'사장 님'+'사장'+'사장님'

def tokenize(doc):
    r = []
    malist = okt.pos(doc, norm=True, stem=True)
    for word in malist:
        if not word[1] in ["Josa", "Eomi", "Punctuation"]:
        # 토큰화 단계에서 필요없는 조사, 접속사, 문장부호 등을 제거
          if not word[0] in stopwords: # 불용어 제거
            string = ''.join([i for i in word[0] if not i.isdigit() and not i.encode().isalpha()])   # 숫자 알파벳 제거
            r.append(string)
        
    rl = (" ".join(r)).strip()
    return rl
```

```python
## 정규표현식
kakao_df.s_review = kakao_df.s_review.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```python
for i in range(len(kakao_df)):
  kakao_df['s_review'][i] = str(kakao_df['s_review'][i]).replace("'","")
```

```python
for i in range(len(kakao_df)):
  kakao_df['s_review'][i] = str(kakao_df['s_review'][i]).split(',')
```

# **word2vec**


```python
total=[] 
for i in range(len(kakao_df['s_review_de'])):
  for j in range(len(kakao_df['s_review_de'][i])):
    total.append(kakao_df['s_review_de'][i][j])
total
```

```python
from gensim.models import Word2Vec
model = Word2Vec(sentences = total, size = 100, window = 5, min_count = 5, workers = 4, sg = 0)
```

# Bokeh

```python
from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row 
from bokeh.plotting import figure
from bokeh.models import HoverTool, ColumnDataSource, value
from bokeh.palettes import brewer

output_notebook()
```

```python
# Bokeh Datasouce 만들기
plot_data = ColumnDataSource(
    data=df.to_dict(orient='list')
)
```

```python
# Plot 만들기(배경)
tsne_plot = figure(
    title='TSNE Embeddings',
    plot_width = 650,
    plot_height = 650,
    active_scroll='wheel_zoom',
    output_backend="webgl",
)
```

```python
# 해당 Hover 툴팁 만들기
tsne_plot.add_tools(
    HoverTool(
        tooltips=[("word", "@word")]
    )
)
```

```python
tsne_plot.circle(
    source=plot_data,
    x='x_coord',
    y='y_coord',
    line_alpha=0.3, 
    fill_alpha=0.2,
    size=10
)
```

```python
# 각 값들 추가해주기 
tsne_plot.title.text_font_size = '16pt'
tsne_plot.xaxis.visible = False
tsne_plot.yaxis.visible = False
tsne_plot.grid.grid_line_color = None
tsne_plot.outline_line_color = None


show(tsne_plot)
```
